```python
import numpy as np
import cv2
import matplotlib
import joblib
import os
import random
```

```python
BASE_DIR = './drive/MyDrive/Minor Project'
COVID_DIR = os.path.join(BASE_DIR, 'Dataset', 'test', 'COVID')
NORMAL_DIR = os.path.join(BASE_DIR, 'Dataset', 'test', 'Normal')
```

```python
covid = os.listdir(COVID_DIR)
normal = os.listdir(NORMAL_DIR)
```

```python
model = joblib.load(os.path.join(BASE_DIR, 'Covid_VGG_16.joblib'))
```

```python
def prediction(images):
  n = len(images)
  ans = []
  for i in range(n):
    img = cv2.imread(os.path.join(COVID_DIR, images[i]), 0)
    img = cv2.resize(img, (256, 256))
    img = img / 255
    img = np.reshape(img, [1, 256, 256, 1])
    res = model.predict(img)
    res = np.argmax(res)
    ans.append('Covid' if(res == 0) else 'Normal')
  return ans
```

```python
images = covid[]
```

```python
prediction(images)
```
